```python
import mysql.connector

# 
config = {
          'user':'root', 
          'password':'', 
          'host':'127.0.0.1', 
          'port':3306,  
          'database':'caiji'}
conn = mysql.connector.connect(**config)

# 
cur = conn.cursor()

# SQL
sql = "SELECT r1,r2,r3,r4,r5,r6 FROM lottery_data"
cur.execute(sql)

# 
result_set = cur.fetchall()
if result_set:
    for row in result_set:
        print ("%d, %d, %d, %d, %d, %d" % (row[0],row[1],row[2],row[3],row[4],row[5]))

#         
cur.close()
conn.close()
```
